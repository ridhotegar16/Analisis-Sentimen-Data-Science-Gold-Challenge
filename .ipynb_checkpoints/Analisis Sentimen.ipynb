```python
def cleaning_text(text): 
    text = re.sub('@[A-Za-z09]+', '', text) 
    text = re.sub('#', '', text)
    text = re.sub('RT[\s]+','', text) 
    text= re.sub('https?:\/\/\S+', '', text)
    return text


data['tweetrapih'] = data['Tweet'].apply(cleaning_text)
data.head()
```

```python
def tokennization(text):
    text = re.split ('\W+', text)
    return text

data['Tokenisasi'] = data['tweetrapih'].apply(lambda x: tokennization(x.lower()))
data.head(10)
```

```python
data['text_length'] = data['tweetrapih'].apply(len)

import matplotlib.pyplot as plt
plt.hist(data['text_length'], bins=30)
plt.xlabel('Panjang Teks')
plt.ylabel('Frekuensi')
plt.title('Distribusi Panjang Teks')
plt.show()
```

```python
def analyze_sentiment_textblob(text):
    analysis = TextBlob(text)
    # TextBlob memberikan nilai sentimen antara -1 dan 1
    if analysis.sentiment.polarity > 0:
        return 'Positif'
    elif analysis.sentiment.polarity == 0:
        return 'Netral'
    else:
        return 'Negatif'

data['sentiment'] = data['tweetrapih'].apply(analyze_sentiment_textblob)

data.head(10)
```

```python
jumlah_positif = data[data['sentiment'] == 'Positif'].shape[0]
jumlah_netral = data[data['sentiment'] == 'Netral'].shape[0]
jumlah_negatif = data[data['sentiment'] == 'Negatif'].shape[0]

print("Jumlah Positif:", jumlah_positif)
print("Jumlah Netral:", jumlah_netral)
print("Jumlah Negatif:", jumlah_negatif)
```

```python
katakasar=pd.read_csv('abusive.csv',encoding='latin1')
katakasar.head()
```

```python
def cek_kata_kasar(kalimat, daftar_kata_kasar):
    kata_dalam_kalimat = set(kalimat.lower().split())
    kata_kasar_ditemukan = kata_dalam_kalimat.intersection(daftar_kata_kasar)
    
    return bool(kata_kasar_ditemukan)

df_kata_kasar = katakasar['ABUSIVE']
df= data
df['katakasar'] = df['tweetrapih'].apply(lambda x: 'Mengandung Kata Kasar' 
                                                         if cek_kata_kasar(x, set(df_kata_kasar)) 
                                                         else 'Tidak Mengandung Kata Kasar')

df
```

```python
jumlah_katakasar = df[df['katakasar'] == 'Mengandung Kata Kasar'].shape[0]
jumlah_katatidakkasar = df[df['katakasar'] == 'Tidak Mengandung Kata Kasar'].shape[0]
print("jumlah katakasar:", jumlah_katakasar)
print("jumlah_kata tidak kasar:", jumlah_katatidakkasar)
```

```python
import pandas as pd
from scipy.stats import chi2_contingency
# Membuat tabel kontingensi
contingency_table = pd.crosstab(df['sentiment'], df['katakasar'])

# Melakukan uji chi-square
chi2, p, _, _ = chi2_contingency(contingency_table)

# Menampilkan hasil uji
print(f"Chi-square value: {chi2}")
print(f"P-value: {p}")

# Menentukan apakah ada hubungan atau tidak
alpha = 0.05
if p < alpha:
    print("Ada hubungan antara sentimen dan kata kasar")
else:
    print("Tidak ada hubungan antara sentimen")
```
